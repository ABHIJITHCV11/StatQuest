```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_rows', None)

# Define the range and create the DataFrame
z = np.arange(-4, 4, 0.1)
df1 = pd.DataFrame(z, columns=['z'])

# Calculate the desired columns
df1["e^(z)"] = np.exp(df1['z'])
df1["e^(-z)"] = np.exp(-df1['z'])
df1["1 + e^(-z)"] = 1 + np.exp(-df1['z'])
df1["1 / (1 + e^(-z))"] = 1 / (1 + np.exp(-df1['z']))
df1
```

```python
print(f"e^1 = {np.exp(1)}")
print(f"e^0 = {np.exp(0)}")
```

```python
# Plotting
plt.figure(figsize=(10, 6))

# Plot each column
plt.plot(df1['z'], df1["e^(z)"], label="e^(z)")
plt.plot(df1['z'], df1["e^(-z)"], label="e^(-z)")
plt.plot(df1['z'], df1["1 + e^(-z)"], label="1 + e^(-z)")
plt.plot(df1['z'], df1["1 / (1 + e^(-z))"], label="1 / (1 + e^(-z))")

# Add labels and title
plt.xlabel('z')
plt.ylabel('Value')
plt.title('Exponential Functions')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
# Plotting e^(z)
plt.figure(figsize=(8, 5))
plt.plot(df1['z'], df1["e^(z)"], label="e^(z)", color='blue')

# Add labels and title
plt.xlabel('z')
plt.ylabel('e^(z)')
plt.title('Plot of e^(z)')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
# Plotting e^(-z)
plt.figure(figsize=(8, 5))
plt.plot(df1['z'], df1["e^(-z)"], label="e^(-z)", color='green')

# Add labels and title
plt.xlabel('z')
plt.ylabel('e^(-z)')
plt.title('Plot of e^(-z)')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
# Plotting 1 + e^(-z)
plt.figure(figsize=(8, 5))
plt.plot(df1['z'], df1["1 + e^(-z)"], label="1 + e^(-z)", color='red')

# Add labels and title
plt.xlabel('z')
plt.ylabel('1 + e^(-z)')
plt.title('Plot of 1 + e^(-z)')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```

```python
# Plotting 1 / (1 + e^(-z))
plt.figure(figsize=(8, 5))
plt.plot(df1['z'], df1["1 / (1 + e^(-z))"], label="1 / (1 + e^(-z))", color='purple')

# Add labels and title
plt.xlabel('z')
plt.ylabel('1 / (1 + e^(-z))')
plt.title('Plot of 1 / (1 + e^(-z))')
plt.legend()

# Display the plot
plt.grid(True)
plt.show()
```
